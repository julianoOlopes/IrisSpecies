```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('Iris.csv')
df
```

```python
df['Species'].value_counts()
```

```python
### Amostragem Aleatória Simples 
```

```python
#utilizando random
import random
indices = random.sample(range(0,149),10)
indices
```

```python
amostra_random = df.iloc[indices]
amostra_random
```

```python
amostra_random['Species'].value_counts()
```

```python
#Utilizando Numpy
import numpy as np
indices_np = np.random.choice(range(0,149),10, replace =False)
indices_np
```

```python
amostra_np = df.iloc[indices_np]
amostra_np
```

```python
amostra_np['Species'].value_counts()
```

```python
#Utilizando pandas
amostra_pd = df.sample(n=10)
amostra_pd
```

```python
amostra_pd['Species'].value_counts()
```

```python
### Amostragem estratificada
```

```python
from sklearn.model_selection import train_test_split
```

```python
_,amostra_sklearn = train_test_split(df, 
                                     stratify= df['Species'],
                                     test_size=0.10)
```

```python
amostra_sklearn['Species'].value_counts()
```
